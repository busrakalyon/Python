```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(2,11,2)
```

```python
y
```

```python
plt.plot(x,y,"red")

plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("X vs Y Graph")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"red")
plt.subplot(2,2,2)
plt.plot(y,x,"blue")
plt.subplot(2,2,3)
plt.plot(y,x,"black")
plt.subplot(2,2,4)
plt.plot(x,y,"yellow")  

# Functional
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,"red")
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X vs Y Graph")

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.2,0.3])

axes2.plot(x,y)
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")

axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")
```

# Subplot ve Figure Özelleştirme

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)
plt.tight_layout()# daha düzgün görünüm için
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)
print(axes) # liste şeklinde olduğu için for döngüsüyle üstünde gezebiliriz
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)
for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)

axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(y,x)
axes[1].set_title("Second Axes")

plt.tight_layout()
```

# Figureleri Özelleştirme

```python
fig = plt.figure(figsize=(4,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2,ncols=1,figsize =(8,8))

axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(y,x**0.5)
axes[1].set_title("Second Axes")
```

```python
fig,axes = plt.subplots(figsize=(8,6))

axes.plot(x,x**0.5,label = "X Karekök",color = "red")
axes.plot(x,x**2,label = "X Kare",color = "blue")
axes.plot(x,x**3,label = "X Küp",color = "#9f7adf")
axes.legend()
```

# Graph Stilleri ve Renkler

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color = "red",linewidth = 5,linestyle = "--")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color = "red",linewidth = 5,linestyle = "-.")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color = "red",linewidth = 5,linestyle = ":")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color = "red",marker ="o",markersize=10
        ,markerfacecolor ="black")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color = "red",marker ="o",markersize=20
        ,markerfacecolor ="black",markeredgecolor = "yellow",markeredgewidth = 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,lw=3,ls ="--")
ax.set_xlim(0,20)
ax.set_ylim(0,50)
```
