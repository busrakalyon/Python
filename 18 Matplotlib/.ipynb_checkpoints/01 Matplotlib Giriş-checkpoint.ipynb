```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

grafikleri çizdirmek için, başka ide'de yazmaya gerek yok

```python
%matplotlib inline 
```

```python
x = np.arange(1,6) # 1'den 6'ya kadar array oluşturur
```

```python
x
```

```python
y = np.arange(2,11,2) #2,11 arası iki atlayarak
```

```python
y
```

```python
#x düzlemi verilerini x arrayinden, y düzlemi verilerini y arrayinden al, grafiğin rengi de kırmızı olsun
plt.plot(x,y, "red") 
plt.show() #bunu idelerde yazmamız lazım ama j.notebookta yazmasak da olur
```

```python
plt.subplot(2,2,1) # küçük küçük grafikler oluşturmak için
plt.plot(x,y,"blue")

#2.grafik
plt.subplot(2,2,2) # küçük küçük grafikler oluşturmak için
plt.plot(y,x,"red")

#3.grafik
plt.subplot(2,2,3) # küçük küçük grafikler oluşturmak için
plt.plot(x,x**2,"yellow")

#4.grafik
plt.subplot(2,2,4) # küçük küçük grafikler oluşturmak için
plt.plot(y,y**2,"purple")

plt.show()
```
